```python
import pandas as pd
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

```python
# Calculate daily returns
df['Return'] = df['Underlying_Price'].pct_change()
```

```python
# Calculate log returns
df['Return'] = np.log(df['Underlying_Price'] / df['Underlying_Price'].shift(1))

# Drop missing values
df = df.dropna()

# Calculate Historical Simulation VaR (95% confidence level)
def calculate_historical_var(returns, confidence_level):
    returns_sorted = np.sort(returns)
    var_index = int((1 - confidence_level) * len(returns_sorted))
    var = returns_sorted[var_index]
    return var

var_95 = calculate_historical_var(df['Return'], 0.95)
print(f"VaR (95% confidence level): {var_95:.4f}")

# Calculate Expected Shortfall (ES)
def calculate_es(returns, confidence_level):
    var = calculate_historical_var(returns, confidence_level)
    es = returns[returns < var].mean()
    return es

es_95 = calculate_es(df['Return'], 0.95)
print(f"Expected Shortfall (95% confidence level): {es_95:.4f}")

# Perform stress testing
stress_test_scenarios = [
    {'name': 'Market Crash', 'return': -0.1},
    {'name': 'Interest Rate Shock', 'return': -0.05},
    {'name': 'Inflation Surge', 'return': -0.07}
]

for scenario in stress_test_scenarios:
    print(f"\nStress Test: {scenario['name']}")
    stressed_returns = df['Return'] + scenario['return']
    var_95_stressed = calculate_historical_var(stressed_returns, 0.95)
    es_95_stressed = calculate_es(stressed_returns, 0.95)
    print(f"VaR (95% confidence level): {var_95_stressed:.4f}")
    print(f"Expected Shortfall (95% confidence level): {es_95_stressed:.4f}")

# Plot histogram of returns with VaR and ES
plt.hist(df['Return'], bins=50, alpha=0.5, label='Returns')
plt.axvline(x=var_95, color='red', label='VaR (95%)')
plt.axvline(x=es_95, color='green', label='ES (95%)')
plt.legend()
plt.show()

```
